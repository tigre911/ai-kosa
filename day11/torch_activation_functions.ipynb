# torch activation functions

## [Non-linear Activations (weighted sum, nonlinearity)](https://pytorch.org/docs/stable/nn.html#non-linear-activations-weighted-sum-nonlinearity)

### Sigmoid (0~1)

```python
import torch
import matplotlib.pyplot as plt

x = torch.linspace(-10, 10, 100)
y = torch.sigmoid(x)

plt.plot(x.numpy(), y.numpy())
plt.show()
```

```python
import torch
from torch import nn

m = nn.Sigmoid()
input = torch.randn(2)
output = m(input)

print(input, output)
```

### Swish

```python
import torch
import torch.nn as nn

class Swish(nn.Module):
    def forward(self, x):
        return x * torch.sigmoid(x)

m = Swish()
input = torch.randn(2)
output = m(input)

print(input, output)
```

```python
import torch
import matplotlib.pyplot as plt

m = Swish()
x = torch.linspace(-10, 10, 100)
y = m(x)

plt.plot(x.numpy(), y.numpy())
plt.show()
```

### Mish

```python
import torch
import torch.nn as nn

class Mish(nn.Module):
    def forward(self, x):
        return x * torch.tanh(torch.nn.functional.softplus(x))

m = Mish()
input = torch.randn(2)
output = m(input)

print(input, output)
```

```python
import torch
import matplotlib.pyplot as plt

m = Mish()
x = torch.linspace(-10, 10, 100)
y = m(x)

plt.plot(x.numpy(), y.numpy())
plt.show()
```

## [Non-linear Activations (other)](https://pytorch.org/docs/stable/nn.html#non-linear-activations-other)

### nn.Softmax (확률)

```python
from torch import nn
import numpy as np

m = nn.Softmax(dim=1)
input = torch.randn(2, 3)
output = m(input)
print(input)
print(output)
print(output.sum(dim=1))
print(np.argmax(output, axis=1)) # 최대값의 인덱스를 찾는 함수. max = 최대값 자체를 찾는다.
```
